```python
df.head()
from sklearn.preprocessing import LabelEncoder
```

```python
nonlabled_df=df.copy()
```

```python
from sklearn import preprocessing
# label_encoder object knows how to understand word labels. 
label_encoder = preprocessing.LabelEncoder()
# Encode labels in column 'Country'. 
df['OFFENSE_CODE']= label_encoder.fit_transform(df['OFFENSE_CODE']) 
print(df.head())
```

```python
df['OFFENSE_TYPE_ID']= label_encoder.fit_transform(df['OFFENSE_TYPE_ID']) 
df['OFFENSE_CATEGORY_ID']= label_encoder.fit_transform(df['OFFENSE_CATEGORY_ID']) 
df['DISTRICT_ID']= label_encoder.fit_transform(df['DISTRICT_ID']) 
df['PRECINCT_ID']= label_encoder.fit_transform(df['PRECINCT_ID']) 
df['VICTIM_COUNT']= label_encoder.fit_transform(df['VICTIM_COUNT']) 
# df['submission_state']= label_encoder.fit_transform(df['submission_state']) 
```
